```python
import pandas as pd
```

```python
df_abe = pd.read_csv('abe_cleaned.csv')
df_suga = pd.read_csv('suga_cleaned.csv')
df_kishida = pd.read_csv('kishida_230713_cleaned.csv')

```

```python
df_merged = pd.concat([df_abe, df_suga,df_kishida])
```

```python
df_merged['publish_date'] = pd.to_datetime(df_merged['publish_date'])
df_sorted = df_merged.sort_values(by='publish_date', ascending=True)
df_sorted.head(3)
```

```python
df_sorted = df_sorted.reset_index(drop=True)
df_sorted = df_sorted.rename(columns={'Unnamed: 0': 'number'})
df_sorted
```

```python
df_sorted.to_csv('PM_activities.csv',encoding='UTF-8')
```
